```python
import matplotlib.pyplot as plt

# Assuming the correct values are already stored in these lists based on the column names
temp_values = farm_data['temperature'].tolist()
humidity_values = farm_data['humidity'].tolist()
moisture_values = farm_data['soil_moisture'].tolist()
rain_values = farm_data['rain'].tolist()
pump_values = [1 if status == 'ON' else 0 for status in farm_data['pump']]

# Plotting the environmental parameters and pump status
plt.figure(figsize=(12, 6))

# Plot the environmental parameters (Temperature, Humidity, and Soil Moisture)
plt.subplot(2, 1, 1)
plt.plot(temp_values, label='Temperature (°C)', color='red')
plt.plot(humidity_values, label='Humidity (%)', color='blue')
plt.plot(moisture_values, label='Soil Moisture (%)', color='orange')
plt.title('Environmental Parameters Over Cycles')
plt.xlabel('Cycle')
plt.ylabel('Values')
plt.legend()

# Plot the pump status (1 for ON, 0 for OFF)
plt.subplot(2, 1, 2)
plt.plot(pump_values, label='Pump Status (1=ON, 0=OFF)', color='green')
plt.title('Irrigation Control Over Time')
plt.xlabel('Cycle')
plt.ylabel('Pump ON/OFF')
plt.legend()

# Adjust layout for a clean display
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import time
from IPython.display import display, Markdown, clear_output, Image

def run_offline_dataset_simulation(df, max_cycles=10):  # Set default to 10 cycles
    temp_values = df['temperature'].tolist()
    humidity_values = df['humidity'].tolist()
    moisture_values = df['soil_moisture'].tolist()
    rain_values = df['rain'].tolist()
    pump_values = [1 if status == 'ON' else 0 for status in df['pump']]

    # Use local images
    pump_on_image_path = "path_to_local_image/Irrigation_Sprinkler.jpg"  # Update path
    pump_off_image_path = "path_to_local_image/Crops_without_irrigation.jpg"  # Update path

    for i in range(min(max_cycles, len(df))):  # Loop only up to max_cycles
        clear_output(wait=True)
        display(Markdown(f"### Cycle {i+1}"))
        display(Markdown(f"- **Temperature**: {temp_values[i]} °C"))
        display(Markdown(f"- **Humidity**: {humidity_values[i]} %"))
        display(Markdown(f"- **Soil Moisture**: {moisture_values[i]} %"))
        display(Markdown(f"- **Rain**: {'Yes' if rain_values[i] == 1 else 'No'}"))
        display(Markdown(f"- **Pump Status**: {'ON' if pump_values[i] == 1 else 'OFF'}"))

        # Display appropriate image (use local image paths)
        if pump_values[i] == 1:
            display(Image(pump_on_image_path))
        else:
            display(Image(pump_off_image_path))

        time.sleep(2)  # Wait 2 seconds before next cycle

    # Final Graphs
    plt.figure(figsize=(12, 6))

    plt.subplot(2, 1, 1)
    plt.plot(temp_values[:max_cycles], label='Temperature (°C)', color='red')
    plt.plot(humidity_values[:max_cycles], label='Humidity (%)', color='blue')
    plt.plot(moisture_values[:max_cycles], label='Soil Moisture (%)', color='orange')
    plt.title('Environmental Parameters Over Cycles')
    plt.xlabel('Cycle')
    plt.ylabel('Values')
    plt.legend()

    plt.subplot(2, 1, 2)
    plt.plot(pump_values[:max_cycles], label='Pump Status (1=ON, 0=OFF)', color='green')
    plt.title('Irrigation Control Over Time')
    plt.xlabel('Cycle')
    plt.ylabel('Pump ON/OFF')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Run the simulation with a limited number of cycles (for example, 10)
run_offline_dataset_simulation(farm_data, max_cycles=10)
```
